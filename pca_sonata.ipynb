```python
import numpy as np 
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/Sonar.csv")
df.head()
```

```python
X_scaled=df.values[:,:-1]
X_scaled
```

```python
features = X_scaled.T
cov_matrix = np.cov(features)
cov_matrix
```

```python
values, vectors = np.linalg.eig(cov_matrix)
```

```python
tot = sum(values)
var_exp = [(i / tot)*100 for i in sorted(values, reverse=True)]
var_exp
```

```python
np.cumsum(var_exp)
```

```python
P = vectors.T.dot(features)
print(P.T)
```

```python
pd.DataFrame(P.T,columns=var_exp)
```
